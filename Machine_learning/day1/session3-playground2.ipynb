```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from numpy import genfromtxt
iris = genfromtxt('iris_data.csv', delimiter=',')
```

## TASK 1
Make an histogram of sepal length for:
- the iris setosa in red,
- the iris versicolor in blue,
- the iris virginica in green.

がくの長さをヒストグラムを作成してください。
- ヒオウギアヤメを赤い線で
- ブルーフラッグを青い線で
- バージニカを緑色の線で

```python
## ヒオウギアヤメのデータのみ抽出
setosa = iris[iris[:,0] == 0, :]
## ブルーフラッグのデータのみ抽出
versicolor = iris[iris[:,0] == 1, :]
## バージニカのデータのみ抽出
virginica = iris[iris[:,0] == 2, :]
## sepal length, sepal width, petal length, petal width に対応する列番号
SL = 1
SW = 2
PL = 3
PW = 4

plt.hist(setosa[:, SL], bins = np.linspace(4, 8, 10), histtype='step', color="r", label="iris setosa")
plt.hist(versicolor[:, SL], bins = np.linspace(4, 8, 10), histtype='step', color="b", label="iris versicolor")
plt.hist(virginica[:, SL], bins = np.linspace(4, 8, 10), histtype='step', color="g", label="iris virginica")
plt.xlabel("length [cm]")
plt.ylabel("count")
plt.title("Histogram of sepal length for the iris setosa(red), the iris versicolor(blue), and the iris virginica(green)")
```

同じプロットに2つ以上のグラフをプロットすると便利な場合があります。そういう時は`subplot`関数が使えます。
<br>
その関数の使用法を調べ、次のプロットを作成します。
- 左側：ブルーフラッグのがくの幅のプロット。
- 右側：ブルーフラッグのがくの幅のヒストグラム。

```python
plt.figure(figsize=(10,5))
# 左側：ブルーフラッグのがくの幅のプロット。
plt.subplot(121)
plt.plot(versicolor[:, SW])
plt.xlabel("samples")
plt.ylabel("cm")
plt.title("Sepal width for the iris versicolor")
 
# 右側：ブルーフラッグのがくの幅のヒストグラム。
plt.subplot(122) 
plt.hist(versicolor[:, SW], bins = np.linspace(1, 4, 10))

plt.xlabel("cm")
plt.ylabel("count")
plt.title("Histogram of sepal width for the iris versicolor")
```
