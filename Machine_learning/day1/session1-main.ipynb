# Jupyter
__Jupyter__  lets the user interactively code and run programs directly in a web browser. 
<br>
The Jupyter architecture is depicted in Fig. 1.

__Jupyter__ を使用すると、プログラムのコーディングや実行を直接ウェブブラウザーで行うことができます。
<br>
Jupyterの構造を図1に示します。

<img src="img/jupyter.png" width="500" height="200">

<center>Fig. 1 The Jupyter architecture </center>

The web browser communicates with the __notebook server__, which delegates the execution of the program to the __kernel__. (In this practice we use Python, so the kernel is a Python kernel.)
<br>
The program is contained in a __notebook__ file (with extension _.ipynb_). 

Webブラウザーは__notebook server__と通信します。プログラムの実行が__kernel（カーネル）__に委任されます。(この実習ではPython言語を使用するため、カーネルはPythonカーネルです。）
<br>
プログラムは、ノットブックファイル（拡張子_.ipynb_）に保存されています。

## Notebook / ノートブック

A __notebook__ is composed of consecutive __cells__ that can contain:
- code,
- formatted text (text and mark-up language like html).

It is possible to __run__ a cell:
- Running a code cell executes the code.
- Running a formatted text cell displays the formatted text.

__ノートブック__は連続した__セル__で構成されています。セルには以下のいずれかを入れます。
- コード
- フォーマットされたテキスト（HTMLのようなテキストやマークアップ言語）。

セルを__実行__することができます：
- コードセルを実行すると、コードが実行されます。
- テキストセルを実行すると、フォーマットされたテキストが表示されます。

### Working with cells / セルの操作

A Jupyter notebook has __two modes__:
- __Edit mode__: when typing in a cell. It is indicated by a green cell border with a green left margin. 
- __Command mode__: when performing notebook level actions. Indicated by a blue cell border with a blue left margin.

Jupyterノートブックには、__ 2つの操作モード__があります。
- __編集モード__：セルに入力するときに使います。セルが緑の線で囲まれています。
- __コマンドモード__：ノートブックレベルのアクションを実行する場合。セルのマージンは青いです。

Many useful commands to interact with the cells are available using buttons at the top of the page.
<br>
Figure 3 below explains what each button does.

セルを操作するためには、ページの上部にあるボタンを利用できます。
<br>
下の図3は、各ボタンの機能を説明しています。

<img src="img/jupyter_tools.png" alt="Tools" style="width: 600px;"/>

<br>
<center>Fig. 3 The command menu</center>

You can also use keyboard shortcuts. Below are some useful commands.
キーボードショートカットを使用することもできます。以下は、いくつかの便利なコマンドです。

Changing mode and running a cell / モードの切り替えとセルの実行:
- To go into command mode use / コマンドモードに入る `Esc`
- To go into edit mode use / 編集モードに入る `Enter`
- To run a cell use / セルの実行 `Ctrl-Enter` or `Shift-Enter` (also moves to next cell)

In command mode / コマンドモードに入っている場合:
- Create a cell above / 上にセルを作成 `a`
- Create a cell bellow / 下にセルを作成 `b`
- Copy a cell / セルのコピー `c`
- Paste a cell / セルの貼り付け `v`
- Delete a cell / セルの削除 `dd`
- Change the cell type to code / セル種類をコードセルに変更 `y`
- Change the cell type to text / セル種類をテキストセルに変更 `m`

In edit mode:
- Copy, cut, paste / コピー、カット、貼り付け `Ctrl-C`, `Ctrl-X`, `Ctrl-V`
- Undo / 元に戻す `Ctrl-z`
- Go to the start / 最初に移動 `Ctrl-home`
- Go to the end / 最後に移動 `Ctrl-end`
- etc. / など

You can find the complete list of keyboard shortcuts in the `Help` menu. 
<br>
全てのキーボードショートカットのリストは `Help` メニューにあります。

#### Kernel / カーネル

The code in the Jupyter notebook is running on a _kernel_.
<br>
The kernel keeps track of the assigned variables and executes the computations.

A notebook with a running kernel appears in green notebook icon in the _dashboard_. (Take a look at the dashboard to see.)
<br>
Moreover, if you select the "Running" tab in the dashboard, you can also see a list of notebooks with running kernels.

In the `Kernel` menu, you have functions to `Restart` or `Shutdown` the current kernel.
<br>
These functions are usefull for re-starting the calcualtions anew or freeing the computation resources.


Jupyterノートブックのコードは_カーネル_で実行されています。
<br>
カーネルは割り当てられた変数を追跡し、計算を実行します。

カーネルが実行されているノートブックは、_ダッシュボード_で緑色のノートブックアイコンで表示されます。（ダッシュボードを見てください。）
<br>
また、ダッシュボードで「Running」タブを選択すると、実行中のカーネルを含むノートブックのリストが表示されます。

`Kernel`メニューには、現在のカーネルを`Restart`（再起動）または `Shutdown`（シャットダウン）する機能があります。
<br>
これらの関数は、計算を新たに再開したり、計算リソースを解放したりするのに役立ちます。

#### Closing and stopping a notebook / ノートブックを閉じて停止する

If you just close the browser page with the notebook, the page is no longer displayed. However, the kernel will still be running and using resources like memory!

There are several ways to make sure you stopped the notebook kernel:

ブラウザページを閉じると、ノートブックのページは表示されなくなります。ただし、カーネルは引き続き実行され、メモリなどのリソースを使用します！

ノートブックカーネルを確実に停止するには、いくつかの方法があります。

- To close the page use `File`->`Close and Halt` from the menu. This will both stop the kernel and close the page.

- Shutdown the kernel (`Kernel->Shutdown`) before closing the page.

- Shutdown the notebook kernel from the dashboard - select checkbox next to the notebook name and then click "Shutdown" on the top.

- ページを閉じるには、メニューから`File`->`Close and Halt`を使用します。これにより、カーネルが停止し、ページが閉じます。

- ページを閉じる前にカーネルをシャットダウンします（ `Kernel-> Shutdown`）。

- ダッシュボードからノートブックカーネルをシャットダウンします。ノートブック名の横にあるチェックボックスを選択し、上部にある「Shutdown」をクリックします。

(This is not so important for day1, but it will become important on day2, especially if you use graphic cards.)

（これは1日目ではそれほど重要ではありませんが、2日目にグラフィックカードを使用する場合がありますので、その時に使用していないカーネルを停止するのが重要になります。）
