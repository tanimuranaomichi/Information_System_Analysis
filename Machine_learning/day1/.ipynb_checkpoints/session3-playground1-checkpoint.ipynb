## TASK 1
Plot the petal length for the iris versicolor and the iris virginica.
- Use the `plot` function to create the plot
- Add the adequate labels

ブルーフラッグ種とバージニカ種の花びらの長さをプロットしてください。
- 関数`plot`を使用してプロットを作成する
- 適切なラベルを追加する


```python
## ヒオウギアヤメのデータのみ抽出
setosa = iris[iris[:,0] == 0, :]
## ブルーフラッグのデータのみ抽出
versicolor = iris[iris[:,0] == 1, :]
## バージニカのデータのみ抽出
virginica = iris[iris[:,0] == 2, :]
## sepal length, sepal width, petal length, petal width に対応する列番号
SL = 1
SW = 2
PL = 3
PW = 4

plt.plot(versicolor[:, PL], "b") 
plt.plot(virginica[:, PL], "g") 
plt.xlabel("samples")
plt.ylabel("cm")
plt.title("Petal length for the iris versicolor(blue) and the iris virginica(green)")
```
